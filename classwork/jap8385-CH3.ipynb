```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np

def montecarlo_pi(numbershots):
    counter = 0
    for i in range (0, numbershots):
        x =random.uniform(-1,1)
        y =random.uniform(-1,1) 
        if (x*x + y*y < 1):
            counter = counter +1
        
    return 4*counter/numbershots
```

```python
relative_errors = []
true_errors = []
true_pi = np.pi

for i in range (0,100):
    calculated_pi=montecarlo_pi(1000)
    error = abs(calculated_pi - true_pi)
    true_errors.append(error)
    relative = abs((true_pi - calculated_pi) / true_pi)
    relative_errors.append(relative)
    print(calculated_pi, error, relative)
fig1 = plt.figure()
plt.plot(range (0,100), true_errors,'r', label = "Absolute/True error")

plt.legend()
plt.show

fig2 = plt.figure()
plt.plot(range (0,100), relative_errors, label="Relative eroor")
plt.legend()
plt.show()
```
