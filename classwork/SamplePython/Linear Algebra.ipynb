```python
#Ax = b  A id A is the coefficient matrix,x is the vector of unknowns, b is the constant vector
#2x +3y = 5
#4x -1y = 3

#Coefficient matrix A and constant vector b

A = np.array([[2,3], [4,-1]])
b = np.array([5,3])
print("A = ",A)
print("b =", b)

#Solving for A * x = b
x=np.linalg.solve(A, b)
print("solution = ",x)
```

```python
E = np.array([[1,2], [2,1]]A
eigenvalues, eigenvectors = np.linalg.eig(E)
print("Eigenvalues = ",eigenvalues)
print("eigenvectors = ", eigenvectors)
```

```python
#Inverse
A= np.array([[4,7], [2,6]])
A_inv = np.linalg.inv(A)
print(A)
print("Inverse of matrix A = ", A_inv)


```

```python
identity = np.dot(A, A_inv)
print(identity)
```

## Singular Value Decomposition (SVD) is a factorization of a matrix into three matrices:
## A=UΣV T
## Where:
## ● U and V are orthogonal matrices,
## ● Σ is a diagonal matrix of singular values.

```python
#Singular Value Decomposition
# Define a matrix
A = np.array([[1, 2], [3, 4], [5, 6]])
# Perform SVD
U, sigma, Vt = np.linalg.svd(A)
print("U matrix:\n", U)
print("Sigma values:", sigma)
print("V^T matrix:\n", Vt)
```
