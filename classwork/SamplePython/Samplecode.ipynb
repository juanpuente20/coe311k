Monte pi

```python
import math
import random
import numpy as np

def montecarlo_pi(numbershots):
    counter = 0
    for i in range (0, numbershots):
        x =random.uniform(-1,1)
        y =random.uniform(-1,1) 
        if (x*x + y*y < 1):
            counter = counter +1
        
    return 4*counter/numbershots
```

```python
for i in range (0,100):
    calculated_pi=montecarlo_pi(1000)
    true_pi=np.pi
    true= abs(calculated_pi - true_pi)
    relative= abs((true_pi - calculated_pi)/true_pi)
    
    print(calculated_pi, true, relative)
    
```
