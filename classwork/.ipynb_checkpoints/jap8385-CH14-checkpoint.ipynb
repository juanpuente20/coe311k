```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its true derivative
def f(x):
    return np.sin(x)

def true_derivative(x):
    return np.cos(x)

# Setup the domain parameters
a = 0
b = 2 * np.pi
N = 30  # Number of points (kept small to visually see the approximation error)

# Step size
h = (b - a) / (N - 1)
x = np.linspace(a, b, N)

# Calculate the True Derivative for comparison
true_dy = true_derivative(x)

# Central Difference Approximation for the derivative
central_diff = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
forward_diff = (f(x + h) - f(x)) / h

# Backward Difference Approximation
backward_diff = (f(x) - f(x - h)) / h

# Plot the approximation and the true derivative
# Create a canvas of 4 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Approximating Derivative of f(x) = sin(x)", fontsize=16)

# 1. Top Left: True Derivative
axs[0, 0].plot(x, true_dy, 'k-', linewidth=2, label='True: cos(x)')
axs[0, 0].set_title('True Derivative')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel("f'(x)")
axs[0, 0].legend()
axs[0, 0].grid(True)

# 2. Top Right: Forward Difference vs True
axs[0, 1].plot(x, true_dy, 'k--', alpha=0.4, label='True: cos(x)') 
axs[0, 1].plot(x, forward_diff, 'r-', label='Forward Diff')
axs[0, 1].set_title('Forward Difference')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel("f'(x)")
axs[0, 1].legend()
axs[0, 1].grid(True)

# 3. Bottom Left: Backward Difference vs True
axs[1, 0].plot(x, true_dy, 'k--', alpha=0.4, label='True: cos(x)')
axs[1, 0].plot(x, backward_diff, 'b-', label='Backward Diff')
axs[1, 0].set_title('Backward Difference')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel("f'(x)")
axs[1, 0].legend()
axs[1, 0].grid(True)

# 4. Bottom Right: Central Difference vs True
axs[1, 1].plot(x, true_dy, 'k--', alpha=0.4, label='True: cos(x)')
axs[1, 1].plot(x, central_diff, 'g-', label='Central Diff')
axs[1, 1].set_title('Central Difference')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel("f'(x)")
axs[1, 1].legend()
axs[1, 1].grid(True)


plt.show()
```
