```python
import numpy as np
import matplotlib.pyplot as plt


A = np.array([[0, 1],
              [-2, -3]])

# Initial condition
x0 = np.array([1, 0])

# Time span
t_final = 10

# Step sizes to test
step_sizes = [0.5, 0.2, 0.1, 0.05]


for h in step_sizes:
    print(f"Step size: {h}")

    # Euler coefficient matrix
    I = np.eye(A.shape[0])
    M = I + h * A

    # Eigenvalues of update matrix
    eigenvalues = np.linalg.eigvals(M)
    magnitudes = np.abs(eigenvalues)

    print("Eigenvalue magnitudes:", magnitudes)

    # Stability check
    if np.all(magnitudes < 1):
        print("Solution is Stable")
    else:
        print("Solution may not be Entirely Stable ;)")

   
    t_values = np.arange(0, t_final + h, h)
    x_values = np.zeros((len(t_values), len(x0)))
    x_values[0] = x0

    for i in range(1, len(t_values)):
        x_values[i] = x_values[i-1] + h * (A @ x_values[i-1])

    
    plt.figure()
    plt.plot(t_values, x_values[:, 0], label='x1(t)')
    plt.plot(t_values, x_values[:, 1], label='x2(t)')
    plt.title(f"Euler Method Solution (h = {h})")
    plt.xlabel("Time")
    plt.ylabel("Solution")
    plt.legend()
    


plt.show()
```
