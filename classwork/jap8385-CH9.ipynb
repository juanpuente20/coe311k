```python
import numpy as np
import matplotlib.pyplot as plt

# Data points 
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")
```

```python
plt.scatter(X,Y, label='Data points')
x = np.linspace(0, 6, 100)
plt.plot(x, (m*x + c),"r", label=f'Best fit line')
plt.title('Best fit line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
np.random.seed(0)   

# Generate random data
X = np.random.rand(10) * 10   # 20 random x values between 0 and 10
Y = np.random.rand(10) * 10   # 20 random y values between 0 and 10

# Sort X (and reorder Y to match X)
sorted_indices = np.argsort(X)
X = X[sorted_indices]
Y = Y[sorted_indices]

# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")
```

```python
plt.scatter(X,Y, label='Data Random')
x = np.linspace(min(X), max(X), 100)
plt.plot(x, (m*x + c),"r", label=f'Best fit line')
plt.title('Best fit line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

It does not fit as good now

```python
m_true = 3
c_true = 2

# x values
x = np.arange(0, 1.01, 0.01)

# Add noise
noise = np.random.normal(0, 0.2, len(x))

# Generate noisy y
y = m_true * x + c_true + noise

A = np.vstack([x, np.ones(len(x))]).T
m_est, c_est = np.linalg.lstsq(A, y, rcond=None)[0]

print(f"Slope: {m_true}, Estimated slope: {m_est}")
print(f"Intercept: {c_true}, Estimated intercept: {c_est}")
```

```python
plt.scatter(x, y, s=10, label="Data")

# True line
plt.plot(x, m_true*x + c_true, 'g', label="True line")

# Estimated line
plt.plot(x, m_est*x + c_est, 'r', label="Least squares fit")

plt.legend()
plt.show()
```
