# Truncation Error 

```python
import numpy as np

true_pi = np.pi

sum_series = 0

for i in range(0,5001):
    term = ((-1)**i) / (2*i + 1)
    sum_series += term
    
    pi_calculated = 4 * sum_series
    relative_error = abs((true_pi - pi_calculated) / true_pi)
    
    if (i) % 100 == 0:
        print( "On the term", i, " therelative error =" ,relative_error)
```

As each term is added to the summation the error gets smaller
