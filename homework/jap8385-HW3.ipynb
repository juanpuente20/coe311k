# Truncation Error Homework 3

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def taylor_sin(x, n_terms):
    term = x          # first term (k = 0)
    total = term

    for i in range(1, n_terms):
        term = term * (-i**2) / ((2*i)*(2*i+1))
        total += term

    return total


# x values from 0 to 10
x_values = np.arange(0, 10.1, 0.1)
sin_exact = np.sin(x_values)

errors_1 = []
errors_2 = []
errors_3 = []
errors_4 = []

for x in x_values:
    approx1 = taylor_sin(x, 1)
    approx2 = taylor_sin(x, 2)
    approx3 = taylor_sin(x, 3)
    approx4 = taylor_sin(x, 4)

    exact = np.sin(x)

    errors_1.append(abs(exact - approx1))
    errors_2.append(abs(exact - approx2))
    errors_3.append(abs(exact - approx3))
    errors_4.append(abs(exact - approx4))


# Plot truncation error
plt.plot(x_values, errors_1, label="1 term")
plt.plot(x_values, errors_2, label="2 terms")
plt.plot(x_values, errors_3, label="3 terms")
plt.plot(x_values, errors_4, label="4 terms")

plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.title("Truncation Error")
plt.legend()
plt.show()
```

If we more terms the truncation error decreases but it is not complety uniformal for all values of x and it workds better with lower x values
