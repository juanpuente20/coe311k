```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np

def montecarlo_pi(numbershots):
    counter = 0
    for i in range (0, numbershots):
        x =random.uniform(-1,1)
        y =random.uniform(-1,1) 
        if (x*x + y*y < 1):
            counter = counter +1
        
    return 4*counter/numbershots
```

```python
samples = list(range(10, 1001, 20))
true_errors = []
true_pi = np.pi

for n in samples:
    calculated_pi = montecarlo_pi(n)
    error = abs(calculated_pi - true_pi)
    true_errors.append(error)
    relative = abs((true_pi - calculated_pi) / true_pi)
    print(calculated_pi, error, relative)

plt.plot(samples, true_errors)
plt.show()
```
