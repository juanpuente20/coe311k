```python
import numpy as np
import matplotlib.pyplot as plt

# Exact solution
def y_exact(x):
    return np.exp(-2*np.sin(x))

# Euler method
def euler_method(f, x0, y0, h, x_end):
    x_vals = np.arange(x0, x_end + h, h)
    y_vals = np.zeros(len(x_vals))
    y_vals[0] = y0

    for i in range(len(x_vals) - 1):
        y_vals[i+1] = y_vals[i] + h * f(x_vals[i], y_vals[i])

    return x_vals, y_vals

# Differential equation
def f(x, y):
    return -2*y*np.cos(x)

# Step sizes
h_values = [0.01, 0.05, 0.15, 0.2, 0.25, 0.5, 1]

x0 = 0
y0 = 1
x_end = 10*np.pi

avg_errors = []
local_errors = []




for h in h_values:
    x_euler, y_euler = euler_method(f, x0, y0, h, x_end)
    y_true = y_exact(x_euler)

    plt.plot(x_euler, y_euler, label=f"h={h}")

    # Average error
    avg_error = np.mean(np.abs(y_true - y_euler))
    avg_errors.append(avg_error)

    # Local error near x ≈ 15
    idx = np.argmin(np.abs(x_euler - 15))
    local_error = abs(y_true[idx] - y_euler[idx])
    local_errors.append(local_error)

# Plot exact solution
x_fine = np.linspace(0, x_end, 1000)
plt.plot(x_fine, y_exact(x_fine), 'k--', label="Exact")

plt.title("Euler Method vs Exact Solution")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()



plt.plot(h_values, avg_errors, 'o-')
plt.title("Average Error vs Step Size h")
plt.xlabel("h")
plt.ylabel("Average Error")
plt.grid()
plt.show()


plt.plot(h_values, local_errors, 's-')
plt.title("Local Error at x = 15 vs Step Size h")
plt.xlabel("h")
plt.ylabel("Local Error")
plt.grid()
plt.show()
```

We can see how at smaller step size and smaller value of x the aprroximation is closer to the actual value, these can be produce because in hight values the error accumulates and only increases
